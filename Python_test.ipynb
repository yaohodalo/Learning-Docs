```python
#import pandas to be able to work with data frames
import pandas as pd
```

```python
#read csv file with pandas
df = pd.read_csv('/SPARK/data/studentData.csv')
df.head(10)
```

```python
#len(df) or df.shape[0] gets the number of rows. df.count() gets the number of rows for each columns
len(df)
```

```python
#df.columns gets the columns's names
df.columns
```

```python
#create a new column total marks = 120
df.head(10)
#delete columns using df.drop(columns=['a','b'])
df.drop(columns=['new_age', 'Total Marks'])
```

```python
#create a new column total marks 120
df['Total Marks'] =120
df.head()
df.drop(columns=['new_age'])
```

```python
# create a new column average to calculate the average marks of the student. (marks/total marks)*100
df['Average']= (df['marks']/df['Total Marks']*100)
df.head(10)
df2 = df.drop(columns=['new_age', 'average'])
df2.head(10)
```

```python
# filter out all the students who have achieved more than 80% marks in OOP course and save it as new df
#pd[columns names] is new columns
#df[df[coulmn name]condition]
#df[(df[column name]condition1) &and |or (df[column2]condition2)]
df_oop =df2[(df2['course']=='OOP') & (df2['Average']>80)]
df_oop
```

```python
#filter out all the students who have achieved more than 60% marks in Cloud course and save it as new df
df_cloud =df[(df['course']=='Cloud') & (df['Average']>60)]
df_cloud.head(10)
df_cloud.drop(columns=['new_age','average'])
```
